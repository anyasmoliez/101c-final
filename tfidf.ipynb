```python


IMDB = pd.read_csv('final_data.csv')
IMDB.drop('Unnamed: 0', axis=1, inplace= True)



IMDB.head(10)
```

```python
X = IMDB['review']
y = IMDB['sentiment']
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, \
                                                    random_state=9)
```

```python
with open('positive-words.txt') as f:
    g_lines = f.readlines()
    good = []
    for l in g_lines:
        good_list = l.split('\n')
        good.append(good_list[0]) #list of the positive words
    good_start_index = good.index("a+")
    good = good[good_start_index:]
```

```python
with open('negative-words.txt') as g:
    b_lines = g.readlines()
    bad = []
    for l in b_lines:
        bad_list = l.split('\n')
        bad.append(bad_list[0]) #list of the negative words
    bad_start_index = bad.index("2-faced")
    bad = bad[bad_start_index:]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


lexicon = list(set(good + bad))



# Create TF-IDF features using the combined lexicon
tfidf_vectorizer = TfidfVectorizer(vocabulary=lexicon)
X_train_tfidf = tfidf_vectorizer.fit_transform(x_train)
X_test_tfidf = tfidf_vectorizer.transform(x_test)
tfidf_vectorized_data = pd.DataFrame(
    data=X_train_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out()
)
```
